```python
import pandas as pd
import numpy as np

data = {'sitio': ['rio_subachoque', 'quebrada_honda', 'rio_botello', 'afluente_san_jose','rio_villeta', 'quebrada_cune', 'rio_tobia', 'quebrada_negra','rio_bahamon', 'rio_guayuriba'],
    'oxigeno': [7.8, 6.2, 5.9, 8.3, 6.1, 4.5, 5.2, 7.0, 6.7, 4.8]}

df_oxigeno = pd.DataFrame(data)

def clasificar_oxigeno(valor):
    if valor < 5:
        return "bajo"
    elif valor <= 6:
        return "medio"
    else:
        return "alto"

df_oxigeno['nivel_oxigeno'] = df_oxigeno['oxigeno'].apply(clasificar_oxigeno)
df_oxigeno

# (alerta de riesgo si el ox√≠geno es menor a  5)
df_oxigeno['riesgo_biologico'] = df_oxigeno['oxigeno'].apply(lambda ox: 'alerta' if ox < 5 else 'sin alerta')
df_oxigeno
```

```python
import pandas as pd

df = pd.read_csv('resultado_oxigeno.csv')

# sitios tienen riesgo (alerta)
sitios_con_riesgo = df[df['riesgo_biologico'] == 'alerta']

# Calcular el porcentaje
porcentaje_riesgo = (len(sitios_con_riesgo) / len(df)) * 100

print (f"Porcentaje de sitios con riesgo: {porcentaje_riesgo:.2f}%")

#  nombres de los sitios con riesgo
print("\nSitios con riesgo biol√≥gico:")
for sitio in sitios_con_riesgo['sitio']:
    print( sitio)
```
