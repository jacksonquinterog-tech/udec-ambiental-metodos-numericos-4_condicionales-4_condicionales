```python
import pandas as pd
import numpy as np

df_temp = pd.DataFrame ({'sitio': ['A', 'B', 'C', 'D','E','F', 'G', 'H', 'I','J'],
'temperatura': [18.5,19.4,22.7,20.1,28.3, 25.2, 26.9, 24.4, 27.6, 30.5]})

# Clasificaci√≥n por temperatura
df_temp['categoria_temp'] = np.where(df_temp['temperatura'] > 25, 'Alta', 'Normal')
df_temp['riesgo_termico'] = np.where(df_temp['categoria_temp'] == 'Alta', 'Alto', 'Bajo')

#  porcentaje de riesgo t√©rmico 
sitios_alto = (df_temp['riesgo_termico'] == 'Alto').sum()
total_sitios = len(df_temp)
porcentaje_riesgo = (sitios_alto / total_sitios) * 100

# Mostrar resultado
print("Porcentaje de sitios con riesgo t√©rmico alto:", porcentaje_riesgo, "%")
```
