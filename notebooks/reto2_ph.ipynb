# üß© Reto 2 ‚Äì pH
**Objetivo:** clasificar pH y generar alerta qu√≠mica.

**Datos:** `../data/muestras_calidad_agua.csv`

### Pasos
1. Carga el CSV en `df`.
2. Extrae `sitio` y `pH`.
3. Crea `clasificacion_pH` con tres rangos.
4. Crea `alerta_quimica` (revisi√≥n si pH < 6.5 o pH > 8.5).
5. Filtra `df_alerta` y exporta `resultado_ph.csv`.

> **Reto adicional:** promedio y desviaci√≥n est√°ndar.

```python
import pandas as pd
import numpy as np
 
df = pd.read_csv("../data/muestras_calidad_agua.csv")
print("Datos cargados correctamente:")
df
```

```python
df_ph = df[['sitio', 'pH']]
df_ph
```

```python
def clasificar_ph(valor):
    if valor < 6.5:
        return "Acido"
    elif valor <= 8.5:
        return "Neutro"
    else:
        return "Basico"
        
df_ph['clasificacion_ph'] = df_ph['pH'].apply(clasificar_ph)

df_ph               
```

```python
# alerta_quimica (alerta si pH < 6.5 o pH > 8.5)
df_ph['alerta_quimica'] = np.where((df_ph['pH'] < 6.5) | (df_ph['pH'] > 8.5), 'alerta ph acido', 'ph alcalino')
df_ph
```

```python
df_alerta = df_ph[df_ph['alerta_quimica'] == True]
df_alerta.to_csv("resultado_ph.csv", index=False)
print("\nArchivo 'resultado_ph.csv' exportado con los sitios en alerta.")
```

```python
# promedio y desviacion (mean, significa promedio medio y std, desviacion estandar )
promedio = df['pH'].mean()
desviacion = df['pH'].std()

# (nf) dependiendo la cantidad de cifras decimales a manejar
print(f"Promedio de pH: {promedio:.1f}")
print(f"Desviaci√≥n est√°ndar de pH: {desviacion:.4f}")
```
